```python
#CLOCK SIGNALS

def Potential(M,r):
    return -(G*M)/(r*AU*((c)**2)) #nah

def clock(PHI,dt,omega=1e9,phase=0,A=1):
    """
    CLOCK SIGNAL 
    for dimensionless delay PHI (time t, position x)
        coordtime dt - to integrate over (JD)
        fundamental freq omega
        arb. phase
        arb. amp A
    """
    img = (1+PHI)*omega*dt
    expot = complex(phase,img)
    c_signal = A*(np.exp(expot))
    
    return c_signal


def cc(c1,c2):
    """
    cross-correlate c1&c2
    """
    return c1*np.conjugate(c2)

t = np.arange(0,50,0.2)
phi1 = 3e-10
phi2 = 4e-10

c1_sig = np.zeros([len(t)],dtype=complex)
c2_sig = np.zeros([len(t)],dtype=complex)
xc_sig = np.zeros([len(t)],dtype=complex)

for i in range(len(t)):
    c_i_1 = clock(phi1,t[i])
    c1_sig[i]=c_i_1
    
    c_i_2 = clock(phi2,t[i])
    c2_sig[i]=c_i_2
    
    xc_sig[i]=cc(c_i_1,c_i_2)

plt.figure()
plt.plot(t,c1_sig,label='c1')
plt.plot(t,c2_sig,label='c2')
plt.plot(t,xc_sig,label='xc')
plt.legend()
plt.show()

#FT
f = np.fft.fftfreq(t.shape[-1])
spc1 = np.fft.fft(c1_sig)
spc2 = np.fft.fft(c2_sig)
spxc = np.fft.fft(xc_sig)

plt.figure()
plt.plot(f,spc1,label='c1')
plt.plot(f,spc2,label='c2')
plt.plot(f,spxc,label='xc')
plt.legend()
plt.show()

#NOISE
#for a simple signal
def TESTSIGNAL (x):
    return np.cos(x)

def TESTSIGNAL2 (x):
    return 3*(np.sin(2*x))

!!!!!!!!!!!!!!!!!!!!!
def NOISE (N,n): #+ FREQ DEPENDENCE (draw n values, mult w/ freq)
    #Const N, generate n data in total
    sigma = np.sqrt(N/2)
    n_r = np.random.normal(0,sigma,n)
    n_i = np.random.normal(0,sigma,n)
    noise = np.zeros([n],dtype=complex)
    for i in range(n):
        noise[i] = complex(n_r[i],n_i[i])
    return noise
```

```python
#NOISE GENERATED AS A FUNCTION OF FREQ
def NOISE_freq (N,f): 
    #Const N, f- frequencies
    sigma = np.sqrt(N/2)
    n = len(f)
    n_r = np.random.normal(0,sigma,n)
    n_i = np.random.normal(0,sigma,n)
    noise = np.zeros([n],dtype=complex)
    for i in range(n):
        noise[i] = f[i]*complex(n_r[i],n_i[i])
    return noise
```

```python
#Gaussian
#generate noises for each freq i.e. deltat in freq domain 
noise = NOISE_freq(N_noise,test_freq) 
data_f = osc_sig_f + noise #add noise, freq domain

#ifft
data_t = np.fft.irfft(data_f)

#time-error function x(t) is the difference between expected nominal time and actual normal time
noise_t = data_t - osc_sig 
noise_t2 = np.abs(noise/(2*np.pi*(0.01)))
#noise_t = data_t - osc_sig

plt.figure(figsize=(10,4))
plt.plot(test_t,data_t,label='data with noise')
plt.plot(test_t,osc_sig,label='clean signal')
plt.xlabel("t")
plt.ylabel("signal(t)")
plt.legend()
plt.show()


#avar
#N = 100 #no. x in each tau, i.e. in each tau there are N time err samples
#tau_1 = np.arange(0,200,0.01*N) 
#tau_val = tau_1[1]-tau_1[0]

# def AVAR_est (time, N, noise_time_domain):
#     """
#     N - no. x in each tau, i.e. in each tau there are N time err samples
#     noise_time_domain - 'x' noise
#     note: time[1]-time[0] = sampling frequency
#     """
#     f0 = time[1] - time[0] #sampfreq
#     tau = np.arange(time[0],time[-1],N*f0)
#     tau_val = tau[1] - tau[0]
#     n_tau = len(tau)
#     avar = np.zeros([n_tau])
    
#     #reshape the time error array
#     x_s = np.reshape(noise_time_domain,(len(tau),N))
    
#     for j in range(n_tau):
#         sum_noise = []
#         for k in range(N-3):
#             sum_noise.append(((x_s[j][k]-2*x_s[j][k+1]+x_s[j][k+2]))**2)#/(2*np.pi*f0))**2) 
#             #sum_noise.append(np.absolute(x_s[j][k]-2*x_s[j][k+1]+x_s[j][k+2]))#NORM OR SQUARE OF REAL PART??
#         #avar[j] = sum(sum_noise)/(2*(tau_val**2)*(N-2))
#         avar[j] = sum(sum_noise)/(2*(tau[j]**2)*(N-2))
        
#     return tau, avar


def AVAR_est (time, N, noise_time_domain):
    """
    N - no. x in each tau, i.e. in each tau there are N time err samples - ARRAY
    noise_time_domain - 'x' noise
    note: time[1]-time[0] = sampling frequency
    """
    f0 = time[1] - time[0] #sampfreq
    tau = np.zeros([len(N)])
    for i in range (len(N)):
        tau[i] = N[i]*f0
    avar = np.zeros([len(N)])
    
    for j in range(len(N)):
        #x_s = noise_time_domain
        sum_noise = 0
        x_s = noise_time_domain#[::N[j]]
        for k in range(N[j]-3):
            sum_noise += (x_s[k]-2*x_s[k+1]+x_s[k+2])**2
        avar[j] = sum_noise/(2*(tau[j]**2)*(N[j]-2))
        #print('N=',N[j])
        #print('tau=',tau[j])

    return tau, avar

##############################
N_array = np.arange(10,1000,5)
##############################

#GAUSSIAN NOISE
avar_estimator = AVAR_est(test_t,N_array,noise_t)
test_tau = avar_estimator[0]
test_avar_est = avar_estimator[1]

plt.figure(figsize=(10,4))
plt.plot(test_tau,test_avar_est,'x')
plt.xlabel('tau')
plt.ylabel('avar')
plt.xscale('log')
plt.yscale('log')
plt.show()
```

```python
#coloured estimator

##################
white_freq = 1
##################

#####################################################
def red(f): #red noise - low frequency long time, 1/f noise
    return 1/(f)

def blue(f,white): #blue noise - high frequency short time, f noise
    return ((f)**0.75-(white)**0.75)+red(white)
####################################################

def NOISE_freq_coloured (N,f,white,red,blue): 
    #Const N, f- frequencies
    
    sigma = np.sqrt(N/2)
    n = len(f)
    n_r = np.random.normal(0,sigma,n)
    n_i = np.random.normal(0,sigma,n)
    noise = np.zeros([n],dtype=complex)
    func = np.zeros([n],dtype=complex)
    
    for i in range(n):
        if f[i] <= white:
            noise[i] = np.abs(red(f[i])*complex(n_r[i],n_i[i]))
            func[i] = red(f[i])
        else:
            noise[i] = np.abs(blue(f[i],white)*complex(n_r[i],n_i[i]))
            func[i] = blue(f[i],white)
    
    plt.figure(figsize=(10,4))
    plt.plot(f,noise,'x',label='noise')
    plt.plot(f,func,label='function')
    plt.xlabel('frequency')
    plt.ylabel('noise')
    plt.title('noise power spec scatter')
    plt.yscale('log')
    plt.xscale('log')
    plt.legend()
    plt.show()
    
    return noise, func



n=len(test_freq)  
noise_c = NOISE_freq_coloured(1,test_freq,white_freq,red,blue)[0]
#noise_c = NOISE_freq_coloured(1,test_freq,white_freq,red,blue)[1] #function
noise_c[0] = 0. + 0.j
osc_data_f =  osc_sig_f +noise_c
osc_data_t = np.fft.irfft(osc_data_f)

noise_c_t = (osc_data_t - osc_sig)

plt.figure(figsize=(10,4))
plt.plot(test_freq,osc_data_f)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('f')
plt.ylabel('osc signal freq')
plt.show()

plt.figure(figsize=(10,4))
plt.plot(test_t[1:-1],osc_data_t[1:-1],label='data')
plt.plot(test_t[1:-1],osc_sig[1:-1],label='clean signal')
plt.xlabel('t')
plt.ylabel('osc signal')
plt.legend()
plt.show()


avar_c_estimator = AVAR_est(test_t,N_array,noise_c_t)
test_c_tau = avar_c_estimator[0]
test_c_avar_est  = avar_c_estimator[1]
plt.figure(figsize=(10,4))
plt.plot(test_c_tau[:],test_c_avar_est[:],'x')
plt.xlabel('tau')
plt.ylabel('avar')
plt.yscale('log')
plt.xscale('log')
plt.title('avar')
plt.show()
```

```python
avar_c_int = AVAR_int(test_t,test_freq,N_array,white_freq)
```
